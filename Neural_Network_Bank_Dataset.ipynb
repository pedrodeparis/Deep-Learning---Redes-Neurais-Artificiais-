```python
import pandas as pd
```

```python
# Infos:


# variáveis categóricas:
# personal loan - aceitou o empréstimo na última campanha ---> isso é o target
# securitis account - cliente possui conta de títulos no  banco?
# cd account - cliente tem conta de certificado de depósito?
# online - cliente usa serviços de banco pela net?
# credit card - cliente usa cartão de crédito do banco?



# variáveis contínuas:
# age
# experience
# income
# CCAvg (gasto médio no cartão de crédito)
# Mortage


# variáveis categóricas ordinais:
# family
# education
```

```python
df = pd.read_excel("/content/Bank_Personal_Loan_Modelling.xlsx")
df.head()
```

```python
df = df.drop(columns=["ID"])
```

```python
df.describe()
```

```python
df["Personal_Loan"].value_counts()
```

```python
X = df.drop(columns=["Personal_Loan"])

y = df["Personal_Loan"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_test = pd.DataFrame(X_test_scaled, columns = X_test.columns)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

clf = MLPClassifier(hidden_layer_sizes=(100, 50, 20), activation="logistic")
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
print(f"Acurácia: {accuracy_score(y_test, y_pred)*100}%")
print(f"F1 score: {f1_score(y_test, y_pred)*100}")
```
